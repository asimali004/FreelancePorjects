```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
#reading dataset from github
df3=pd.read_csv("https://raw.githubusercontent.com/fh-aachen-semesterprojekt/Staedteregion-Aachen/main/Pressemitteilung-Entwicklung-2021.csv")
```

```python
df3
```

```python
#selecting all columns except last column
df3=df3[::-1]
#reindexing the dataframe
df3.index=[i for i in range(0,len(df3))]
```

```python
#adjusting figure size in matplotlib
plt.figure(figsize=(16,6))
#adjusting linewidth in matplotlib
plt.plot(df3["Tag"],df3.iloc[:,2],lw=2)
#adjusting font size and colour of the title in matplotlib graph
plt.title("Aktive_Faelle",fontsize=25,color="b")
plt.xlabel("Date")
#for different style available in matplotlib run this command "style.available"
style.use("classic")
#adjusting grid parameters in matplotlib
plt.grid(color="g",ls="-",lw=0.5)
```

```python
plt.figure(figsize=(16,6))
plt.plot(df3["Tag"],df3.iloc[:,3],lw=2)
plt.title("Gesamt",fontsize=25,color="r")
plt.xlabel("Date")
style.use("classic")
plt.grid(color="c",ls="-",lw=0.5)
```

```python
plt.figure(figsize=(16,6))
plt.plot(df3["Tag"],df3.iloc[:,4],lw=2,)
plt.title("Sieben_Tage_Inzidenz",fontsize=25,color="m")
plt.xlabel("Date")
style.use("classic")
plt.grid(color="y",ls="-",lw=0.5)
```
